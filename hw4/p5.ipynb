```python
from keras.models import load_model
import os

lstm_model = load_model(os.path.join(os.getcwd(),'model','LSTM.h5'))
```

```python
lstm_model.summary()
```

```python
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import one_hot, text_to_word_sequence
from keras.preprocessing.text import Tokenizer
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM

import pandas as pd
```

```python

file_name = 'training_label.txt'

train = os.path.join(os.getcwd(),'dataset', file_name)
df_train = pd.read_csv(train,header=None, delimiter='\n')
df_train.columns = ['raw']
df_train['x'] = df_train['raw'].apply(lambda x: x.split(' +++$+++ ')[1])
df_train['y'] = df_train['raw'].apply(lambda x: x.split(' +++$+++ ')[0])

# df_train['cutword'] = df_train['x'].apply(lambda x: text_to_word_sequence(x, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True, split=" "))

df_train

```

```python
import pickle
path = os.path.join(os.getcwd(),'model','tokenizer_lstm.pickle')
tok = pickle.load(open(path, 'rb'))
```

```python
file_name = 'training_nolabel.txt'

path = os.path.join(os.getcwd(),'dataset', file_name)
df_no_label = pd.read_csv(path,header=None, delimiter='\n')
df_no_label.columns = ['raw']
df_no_label['x'] = df_no_label['raw'].apply(lambda x: x)

df_no_label

```

```python
x = df_no_label.x
x_lstm = tok.texts_to_sequences(x)

maxlen = 50
x_lstm = pad_sequences(x_lstm, maxlen=maxlen, padding='pre') 
x_lstm.shape
```

```python
y = lstm_model.predict_classes(x_lstm)
y.shape
```

```python
y = y.reshape(y.shape[0])
y.shape
```

```python
for i in range(10):
    print('------')
    print(df_no_label.x[i])
    print(y[i])
```

```python
x_train = np.concatenate([df_no_label.x, df_train.x], axis=0)
y_train = np.concatenate([y, df_train.y], axis=0)
print(x_train.shape)
print(y_train.shape)
```

```python
x_train = tok.texts_to_sequences(x_train)

maxlen = 50
x_train = pad_sequences(x_train, maxlen=maxlen, padding='pre') 
x_train.shape
```

```python
num_words = 20000

model = Sequential()
model.add(Embedding(num_words, 128))
model.add(LSTM(128, dropout=0.5, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))

model.summary()

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

batch_size=1000
epochs=5

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_split=0.1)
```
