```python
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import one_hot, text_to_word_sequence
from keras.preprocessing.text import Tokenizer
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM

import os
import pandas as pd
```

```python

file_name = 'training_label.txt'

train = os.path.join(os.getcwd(),'dataset', file_name)
df_train = pd.read_csv(train,header=None, delimiter='\n')
df_train.columns = ['raw']
df_train['x'] = df_train['raw'].apply(lambda x: x.split(' +++$+++ ')[1])
df_train['y'] = df_train['raw'].apply(lambda x: x.split(' +++$+++ ')[0])

# df_train['cutword'] = df_train['x'].apply(lambda x: text_to_word_sequence(x, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True, split=" "))

df_train
```

# with punctuation

```python
print('"!#$%&()*+,-./:;<=>?@[\\]^_`{|}~1\t2\n3')
```

```python
num_words = 20000
tok = Tokenizer(num_words=num_words,
                filters='',
                lower=True,
                split=' ', 
                char_level=False)

tok.fit_on_texts(df_train.x)
sorted(tok.word_counts.items(), key=lambda x: x[1], reverse=True)
```

```python
x_train_with_punctuation = tok.texts_to_sequences(df_train.x)
maxlen = 50
x_train_with_punctuation = pad_sequences(x_train_with_punctuation, maxlen=maxlen, padding='pre')

y_train = np.array(df_train.y)
```

# without punctuation

```python
num_words = 20000
tok = Tokenizer(num_words=num_words,
                filters='"!#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
                lower=True,
                split=' ', 
                char_level=False)

tok.fit_on_texts(df_train.x)
sorted(tok.word_counts.items(), key=lambda x: x[1], reverse=True)
```

```python
x_train_without_punctuation = tok.texts_to_sequences(df_train.x)
maxlen = 50
x_train_without_punctuation = pad_sequences(x_train_without_punctuation, maxlen=maxlen, padding='pre')
```

# training

```python
model = Sequential()
model.add(Embedding(num_words, 128))
model.add(LSTM(128, dropout=0.5, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
batch_size=1000
epochs=5

model.fit(x_train_with_punctuation, 
          y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_split=0.25)
```

```python

model.fit(x_train_without_punctuation, 
          y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_split=0.25)
```
